```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.use('MacOSX')
%matplotlib inline
```

```python
# numpy load I_q_IPA_exp.npy
exp = np.load('I_q_IPA_exp.npy')
model = np.load('I_q_IPA_model.npy')
```

```python
x1 = exp[:, 0]
y1 = exp[:, 1]

x2 = model[:, 0]
y2 = model[:, 1]

fig = plt.figure()
ax1  = fig.add_subplot(111)
ax1.plot(x1, y1, 'o', label='data', color='blue')
ax1.plot(x2, y2, '-', label='model', color='red')
plt.show()
```

```python
scaler = 300000/35

y2 /= scaler

fig = plt.figure()
ax1  = fig.add_subplot(111)
ax1.plot(x1, y1, 'o', label='data', color='blue')
ax1.plot(x2, y2, '-', label='model', color='red')
plt.show()
```
